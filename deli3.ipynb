```python
import matplotlib.pyplot as plt

# extract the zipcodes from the original df_2020
df['Zip Code'] = df['LOCATION'].str.split(' ').str[-1]

columns2 = df.loc[(df['CLOSED_DT'].isna() == False)]

columns2['Zip Code'] = df['LOCATION'].str.split(' ').str[-1]
columns = columns2["REASON"]

# group the dataframes by zip code

df_grouped = df.groupby('Zip Code')
columns_grouped = columns2.groupby('Zip Code')

# count the number of rows (i.e., cases) per zip code
num_cases = df_grouped.size()
num_cases_mod = columns_grouped.size()

# calculate the proportion of unresolved cases
prop_unresolved = columns_grouped.size() / df_grouped.size()
prop_unresolved = prop_unresolved.sort_values(ascending = True)
prop_unresolved.plot(kind='bar')

plt.title('Proportion of Resolved vs Total Cases Based on Zip Code')
plt.xlabel('Zip Code')
plt.ylabel('Proportion')
```

```python
# group the dataframes by reason
df_grouped = df.groupby('REASON')
columns_grouped = columns2.groupby('REASON')

# count the number of rows (i.e., cases) per zip code
num_cases = df_grouped.size()
num_cases_mod = columns_grouped.size()

# calculate the proportion of unresolved cases
prop_unresolved = columns_grouped.size() / df_grouped.size()
prop_unresolved = prop_unresolved.dropna()
prop_unresolved = prop_unresolved.sort_values(ascending = True)

prop_unresolved_top20 = prop_unresolved.head(25)
prop_unresolved_top20.plot(kind='bar')

plt.title('Proportion of Resolved vs Total Cases Based on Reason (Top 25)')
plt.xlabel('Reason')
plt.ylabel('Proportion')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

fall = "311_all.csv"
fclosed = "311_closed.csv"


df = pd.read_csv("/Users/aryaan/Desktop/Projects/Analyzing_311_calls/311_block_csv.csv")
closed_data = df[df['CLOSED_DT'].notnull()]
closed_data.to_csv("wack", index=False)
cdf = pd.read_csv("wack")

pdf = cdf.groupby('BLOCK_ID').agg('count')/df.groupby('BLOCK_ID').agg('count')
pdf = pdf[['Unnamed: 0']]
pdf.rename(columns={'Unnamed: 0': 'Proportions'}, inplace = True)
pdf = pdf[pdf['Proportions'].notnull()]

pdf = pdf.reset_index().rename(columns={'index': 'BLOCK_ID'})
pdf['BLOCK_ID'] = pdf['BLOCK_ID'].astype(str).str.slice(stop=-2)
pdf = pdf.sort_values(by='Proportions')
# pdf.to_csv("/content/block.csv", index=False)
plt.bar(pdf['BLOCK_ID'][:25], pdf['Proportions'][:25] )
plt.title('Proportions of top 20 resolved Block Id')
plt.xlabel('Block ID')
plt.xticks(rotation=90)
plt.ylabel('Proportion')
plt.show()
```
